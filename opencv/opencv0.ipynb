```python
import cv2
import numpy as np
```

```python
# 读图片
def img_demo():
    img = cv2.imread('images/000002.jpg') 
    cv2.namedWindow('name',cv2.WINDOW_NORMAL)  # 窗口
    cv2.imshow('name',img)
    #gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)  # 类型转换
    #cv2.imshow('gray',gray)
    # cv2.imwrite("save_path",img)  # 保存的地方
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 图像色域变换   
def color_space_demo():
    img = cv2.imread('images/000002.jpg') 
    cv2.namedWindow('name',cv2.WINDOW_NORMAL)  # 窗口
    #cv2.imshow('name',img)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  # 类型转换
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)    # 机器喜欢
    cv2.imshow('gray',gray)
    cv2.imshow('hsv',hsv)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 一切图像皆是numpy
def mat_demo():
    img = cv2.imread('images/000002.jpg') 
    print(img.shape)
    roi = img[100:200,100:200,:]  # 切小部分图
    blank = np.zeros_like(img)
    blank[100:200,100:200,:] = img[100:200,100:200,:]  # 形状和img一样
    cv2.imshow('img',img)
    cv2.imshow('roi',roi)
    cv2.imshow('blank',blank)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 图像像素操作
def pixel_demo():
    img = cv2.imread('images/000002.jpg') 
    cv2.imshow('img',img)
    h,w,c = img.shape
    for row in range(h):
        for col in range(w):
            b,g,r = img[row,col]
            img[row,col] = (255-b,255-g,255-r)
    cv2.imshow('result',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 像素算术运算 
def math_demo():
    img = cv2.imread('images/000002.jpg') 
    cv2.imshow('img',img)
    h,w,c = img.shape
    blank = np.zeros_like(img)
    blank[:,:] = (50,50,50)
    results = cv2.add(img,blank)      # 加减乘除 ，大小必须一致，数据类型可以不一致
    cv2.imshow('result',results)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def nothing(x):
    print(x)
    
# 滚动条调整图像亮度
def adjust_lightness_demo():
    img = cv2.imread('images/000002.jpg') 
    cv2.namedWindow('input',cv2.WINDOW_NORMAL)
    cv2.createTrackbar('lightness','input',0,100,nothing)
    cv2.imshow('input',img)
    h,w,c = img.shape
    blank = np.zeros_like(img)
    while True:
        pos = cv2.getTrackbarPos('lightness','input')      
        blank[:,:] = (pos,pos,pos)
        results = cv2.add(img,blank)      # 加减乘除 ，大小必须一致，数据类型可以不一致
        cv2.imshow('result',results)
        c = cv2.waitKey(1)
        if c == 27:   # 按esc键关闭
            break
    cv2.destroyAllWindows()
    
# 键盘响应
def key_demo():
    img = cv2.imread('images/000002.jpg') 
    cv2.namedWindow('input',cv2.WINDOW_NORMAL)
    cv2.imshow('input',img)
    h,w,c = img.shape
    blank = np.zeros_like(img)
    while True:
        c = cv2.waitKey(1)  # 视频操作停留1ms
        if c==49: # 1
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            cv2.imshow('gray',gray)
        if c==50: #  2
            hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
            cv2.imshow('hsv',hsv)
        if c == 27:   # 按esc键关闭
            break
    cv2.destroyAllWindows()
    

# 颜色查找表
def color_table_demo():
    colormap = [ 
        # 一些cv2的颜色
        cv2.COLORMAP_AUTUMN,
        cv2.COLORMAP_BONE,
        cv2.COLORMAP_CIVIDIS, 
        cv2.COLORMAP_COOL, 
        cv2.COLORMAP_DEEPGREEN, 
        cv2.COLORMAP_HOT,
        cv2.COLORMAP_HSV, 
        cv2.COLORMAP_INFERNO,
        cv2.COLORMAP_JET,
        cv2.COLORMAP_MAGMA,
        cv2.COLORMAP_OCEAN, 
        cv2.COLORMAP_PARULA, 
        cv2.COLORMAP_PINK, 
        cv2.COLORMAP_PLASMA, 
        cv2.COLORMAP_RAINBOW, 
        cv2.COLORMAP_SPRING, 
        cv2.COLORMAP_SUMMER,
        cv2.COLORMAP_TURBO, 
        cv2.COLORMAP_TWILIGHT,
        cv2.COLORMAP_TWILIGHT_SHIFTED,
        cv2.COLORMAP_VIRIDIS,
        cv2.COLORMAP_WINTER
    ]
    img = cv2.imread('images/cy.png') 
    cv2.namedWindow('input',cv2.WINDOW_NORMAL)
    cv2.imshow('input',img)
    # 循环变换就行
    index = 0
    while True:
        dst = cv2.applyColorMap(img,colormap[index%21])
        index += 1
        cv2.imshow('color strle',dst)
        c = cv2.waitKey(400) 
        #cv2.imwrite('images/cycolor.png',dst)
        if c == 27:   # 按esc键关闭
            break
    cv2.imwrite('images/cycolor.png',dst)
    cv2.destroyAllWindows()
    
# 图像的像素逻辑操作
def bitwise_demo():
    a1 = np.zeros((224,224,3))
    a1[:,:] = (255,0,0)    # 255:白，0：黑
    b1 = np.zeros((224,224,3))
    b1[:,:] = (0,255,255)
    cv2.imshow('colora1',a1)
    cv2.imshow('colorb1',b1)
    
    a2 = cv2.bitwise_and(a1,b1)
    b2 = cv2.bitwise_or(a1,b1)
    cv2.imshow('colora2',a2)
    cv2.imshow('colorb2',b2)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# 图像色彩变换抠图:绿幕抠图
def color_trans_demo():
    img = cv2.imread('images/m1.jpg') 
    cv2.imshow('img',img)
    cv2.namedWindow('img',cv2.WINDOW_NORMAL)  # 窗口
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)    # 机器喜欢
    cv2.imshow('hsv',hsv)
    mask = cv2.inRange(hsv,(35,43,46),(77,255,255))  # 过滤绿色
    cv2.imshow('mask',mask)
    cv2.bitwise_not(mask,mask)    # 取反，把内容变为大于0的数
    results = cv2.bitwise_and(img,img,mask=mask)  # 在mask范围里进行相与
    cv2.imshow('results',results)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 图像像素的统计、
def pixels_stat_demo():
    img = cv2.imread('images/m1.jpg') 
    cv2.imshow('img',img)
    mean,std = cv2.meanStdDev(img)
    print("mean:",mean,"\n","std:",std)  # 怎么检查一张图片是否有有效信息，看std是否有值
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 几何形状的绘制
def draw_demo():
    img = cv2.imread('images/m1.jpg') 
    cv2.rectangle(img,(50,50),(100,100),(0,0,255),4,8,0)
    cv2.circle(img,(50,50),50,(0,255,255),4,8,0)
    cv2.line(img,(50,50),(100,100),(255,0,255),4,8,0)
    cv2.putText(img,'99%',(50,50),cv2.FONT_HERSHEY_SIMPLEX,1.0,(0,255,255),2,8)
    cv2.imshow('img',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 随机数和随机颜色的生成    
def random_demo():
    img = np.zeros([512,512,3],dtype=np.int8)
    while True:
        xx = np.random.randint(0,512,2,dtype=np.int)
        yy = np.random.randint(0,512,2,dtype=np.int)
        bgr = np.random.randint(0,255,3,dtype=np.int)
        # 在一幅图上生成随机的线条
        line = cv2.line(img,(xx[0],yy[0]),(xx[1],yy[1]),(np.int(bgr[0]),np.int(bgr[1]),np.int(bgr[2])),1,8,0)
        cv2.imshow('line',line)
        c = cv2.waitKey(100)
        if c==27:
            break
    cv2.destroyAllWindows()
    
# 轮廓绘制
def polyline_draw_demo():
    canvas = np.zeros((512,512,3),dtype=np.uint8)  # 注意dtype=np.uint8
    cv2.namedWindow('canvas',cv2.WINDOW_NORMAL)  # 窗口
    pts = np.array([[100,100],[350,100],[450,280],[320,450],[80,400]],dtype=np.int32)
    cv2.drawContours(canvas,[pts],-1,(255,0,0),-1)  # -1表示填充
    cv2.imshow('canvas',canvas)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    
    
# 鼠标操作,在图上画一个矩形框
#imgmouse = np.zeros((512,512,3),dtype=np.uint8)
imgmouse = cv2.imread('images/m.jpg')
ori_img = np.copy(imgmouse)
x1 = -1
x2 = -1
y1 = -1
y2 = -1
def mouse_drawing(event,x,y,flags,param):
    global x1,x2,y1,y2
    if event==cv2.EVENT_LBUTTONDOWN:
        #print(x,y)
        x1 = x
        y1 = y
    if event==cv2.EVENT_MOUSEMOVE:
        #print(x,y)
        if x1<0 or y1<0: # 鼠标放下去后才进行
            return
        x2 = x
        y2 = y
        dx = x2-x1
        dy = y2-y1
        if dx>0 and dy>0:
            imgmouse[:,:,:]= ori_img
            cv2.rectangle(imgmouse,(x1,y1),(x2,y2),(0,0,255),2,8,0)
    if event==cv2.EVENT_LBUTTONUP:
        #print(x,y)
        x2 = x
        y2 = y
        dx = x2-x1
        dy = y2-y1
        if dx>0 and dy>0:
            imgmouse[:,:,:]= ori_img
            cv2.rectangle(imgmouse,(x1,y1),(x2,y2),(0,0,255),2,8,0)
        x1 = -1
        x2 = -1
        y1 = -1
        y2 = -1    
        

def mouse_demo():
    cv2.namedWindow('mouse',cv2.WINDOW_AUTOSIZE)
    # 创建一个鼠标回调操作
    cv2.setMouseCallback('mouse',mouse_drawing)
    while True:
        cv2.imshow('mouse',imgmouse)
        c = cv2.waitKey(10)
        if c==27:
            break
    cv2.destroyAllWindows()
    
# 图像像素类型的归一化
'''opecv只能显示np.uint8和浮点型的，float32都不行
   归一化/255.0或者用cv2.normlize()
'''

# 图像变换
def trans_demo():
    img = cv2.imread('images/m.jpg')
    h,w,c = img.shape
    cv2.imshow('ori_img',img)
    resize = cv2.resize(img,(512,512))
    cv2.imshow('resize',resize)
    flip = cv2.flip(img,-1)
    cv2.imshow('flip',flip)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
 # 旋转   
def trans_rotate_demo():
    img = cv2.imread('images/m.jpg')
    h,w,c = img.shape
    cv2.imshow('ori_img',img)
    # 定义矩阵
    M = np.zeros((2,3),dtype=np.float32)
    # 定义角度
    alpha = np.cos(np.pi/4.0)
    beta = np.sin(np.pi/4.0)
    # 初始化矩阵
    M[0,0] = alpha
    M[1,1] = alpha
    M[0,1] = beta
    M[1,0] = -beta
    cx = w/2
    cy = h/2
    tx = (1-alpha)*cx-beta*cy
    ty = beta*cx+(1-alpha)*cy
    M[0,2] = tx
    M[1,2] = ty
    # 执行旋转
    dst = cv2.warpAffine(img,M,(w,h))
    cv2.imshow('rotate',dst)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    

#img_demo() 
#color_space_demo()
#mat_demo()
#pixel_demo()
#math_demo()
#adjust_lightness_demo()
#key_demo()
#color_table_demo()
#bitwise_demo()
#color_trans_demo()
#pixels_stat_demo()
#draw_demo()
#random_demo()
#polyline_draw_demo()
#mouse_demo()
#trans_demo()
trans_rotate_demo()
```

```python
# 读视频
def video_frame():
    cap = cv2.VideoCapture(0)  # 0表示自己的摄像头
    while True:
        ret,frame = cap.read() # 读取视频流
        if ret is not True:
            break
        cv2.imshow("frame",frame)
        c = cv2.waitKey(10)
        if c == 27:
            break
    cap.release() 

video_frame()
```
